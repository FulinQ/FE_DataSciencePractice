# Question 1. Create a function to find Min, Max, Mean, and Median on a list.
Input(parameters):
1. name of the function(String) and
2. List of values

Output:
1. Metric value

PS. sort vs sorted [link text](https://www.geeksforgeeks.org/python-difference-between-sorted-and-sort/)

```python
vals = [7,14,81,56,9,17,43]

def getMetric(metric,ListVal):
   if metric == 'mean':
      return sum(ListVal)/len(ListVal)
   if metric == 'median' :
      ListVal.sort()
      if len(ListVal) % 2 == 1:
         return ListVal[len(ListVal)//2]
      else :
         result = ListVal[len(ListVal)//2] + ListVal[len(ListVal)//2-1]
         return result/2

mean = getMetric("mean",vals)
med = getMetric("median",vals)
print(mean)
print(med)
```

```python
inPath = "sampleText.txt"

with open(inPath) as f: #read as one long string
    contents = f.read()

words = contents.split()
wordCount = dict()
for w in words:
    w = re.sub(r'[^A-Za-z]', '', w)
    wordCount[w] = wordCount.get(w, 0) + 1

wordCount = sorted(wordCount.items(), key=lambda x:(x[1],x[0]))
wordCount = dict(wordCount)

print(wordCount)
for i in wordCount:
    print(i, wordCount[i])
```

```python
inPath = "sampleText.txt"

with open(inPath) as f: #read as one long string
    contents = f.read()

words = contents.split()
wordCount = dict()
for w in words:
    w = re.sub(r'[^A-Za-z]', '', w)
    wordCount[w] = wordCount.get(w, 0) + 1

wordCount = sorted(wordCount.items(), key=lambda x:(x[1]), reverse = True)
wordCount = dict(wordCount)

print(wordCount)
for i in wordCount:
    print(i, wordCount[i])
```
