```python
df_exams = pd.read_csv('StudentsPerformance.csv')
print(df_exams)
```

```python
df_exams['average'] = df_exams['math score']*0.5 + df_exams['reading score']*0.25 +df_exams['writing score']*0.25
print(df_exams)
```

```python
def grader(score):
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'


df_exams['grade'] = df_exams.apply(lambda x: grader(x['average']), axis =1)
print(df_exams)
```

```python
print(df_exams.groupby('grade').size())
print(df_exams['grade'].value_counts().sort_index())
```

```python
print(df_exams.groupby('grade').size()*100/len(df_exams))
print(df_exams['grade'].value_counts().sort_index()*100/len(df_exams))
```

```python
inp = input() #A
fix = 'ABCDF'
if inp in fix:
    print('Number of student getting', inp, '\t\t', len(df_exams[df_exams['grade']==inp]))
    print('Percentage of student getting', inp, '\t', len(df_exams[df_exams['grade']==inp])*100/len(df_exams), '%')
```

```python
selectDf = df_exams[(df_exams['gender'] == 'male') & (df_exams['grade']=="A")]
maleDf = df_exams[df_exams['gender'] == 'male']
print(f'Number of male getting A :\t\t{len(selectDf)}  students')
print(f'Number of male :\t\t\t{len(maleDf)} students')
print(f'Percentage of male getting A to male :\t{round(len(selectDf)*100/len(maleDf),2)} %')
```

```python
aDf = df_exams[df_exams['grade']=="A"]
print(aDf['parental level of education'].value_counts())

assoDf = df_exams[(df_exams['grade']=="A") & (df_exams['parental level of education']=="associate's degree")]
highDf = df_exams[(df_exams['grade']=="A") & (df_exams['parental level of education']=="high school")]

print()
print(len(assoDf) > len(highDf))
```
