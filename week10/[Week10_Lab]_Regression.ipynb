```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
import os

os.chdir('/Users/fulinq/Documents/KMITL/FinancialEngineering/Y2/T2/FE_IntroToDataScience/week10')
df = pd.read_csv("CarPrice_Dataset_GDSC_article.csv")

# !gdown 1QUlM519dvEzQtTSnIzk4CClaL_IopisA

# df = pd.read_csv("/content/CarPrice_Dataset_GDSC_article.csv")
df.head()
```

```python
df = df.drop(['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem'], axis=1)
df = df.drop(['doornumber', 'cylindernumber', 'peakrpm', 'compressionratio', 'peakrpm', 'stroke', 'carheight', 'wheelbase'], axis=1)
# print(df)

x=df.drop(['price'], axis=1)
y=df['price']
# print(x)
# print(y)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)

lr=LinearRegression()
lr.fit(x_train,y_train)

y_predLR = lr.predict(x_test)
# print(y_test)
# print(y_predLR)

resultDf = pd.DataFrame({'y_test': y_test, 'y_predict':y_predLR})
print(resultDf)
print(f'MAE = {metrics.mean_absolute_error(y_test, y_predLR)}')
```

## Task 3
Task 3: Instead of using train_test_split with 0.1 ratios, implement K-fold validation and train the regression model with the same settings.

* Use 5-fold validation and report the average MAE score.

[ Using Kfold article.](https://www.askpython.com/python/examples/k-fold-cross-validation)






```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("CarPrice_Dataset_GDSC_article.csv")
# df = pd.read_csv("/content/CarPrice_Dataset_GDSC_article.csv")

string_feature = [feature for feature in df.columns if df[feature].dtypes=="O"]
one_hot_encoded_data = pd.get_dummies(df, columns = string_feature)
# print(one_hot_encoded_data)

x=one_hot_encoded_data.drop(['price'], axis=1)
y=one_hot_encoded_data['price']
# print(x)
# print(y)

k = 5
kf = KFold(n_splits=k, random_state=None)
model = LinearRegression()

acc_score = []

for train_index , test_index in kf.split(x):
    x_train , x_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_train , y_test = y[train_index] , y[test_index]

    model.fit(x_train,y_train)
    pred_values = model.predict(x_test)

    acc = metrics.mean_absolute_error(y_test,pred_values)
    acc_score.append(acc)

avg_acc_score = sum(acc_score)/k

print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))
```
